```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating three type of dataframe 
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['input1']=[1,1,0,0]
or_data['input2']=[1,0,1,0]
or_data['ouput']=[1,1,1,0]
```

```python
and_data['input1']=[1,1,0,0]
and_data['input2']=[1,0,1,0]
and_data['ouput']=[1,0,0,0]
```

```python
xor_data['input1']=[1,1,0,0]
xor_data['input2']=[1,0,1,0]
xor_data['ouput']=[0,1,1,0]
```

```python
print(and_data)
print(or_data)
print(xor_data)
```

```python
# first try to fit and data using single perceptron
sns.scatterplot(x=and_data['input1'],y = and_data['input2'],hue=and_data['ouput'],s=200)
```

```python
from sklearn.linear_model import Perceptron
clf1=Perceptron()
```

```python
clf1.fit(and_data.iloc[:,0:2],and_data.iloc[:,2])
```

```python
print(clf1.coef_)
print(clf1.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values, clf=clf1, legend=2)
```

So, it can handle and data (as we can draw a linear boundry)

```python
# try to fit or data using single perceptron

clf2=Perceptron()
clf2.fit(or_data.iloc[:,0:2],or_data.iloc[:,-1])

print(clf2.coef_)
print(clf2.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(or_data.iloc[:,0:2].values,or_data.iloc[:,-1].values, clf=clf2, legend=2)
```

```python
# try to fit xor data using single perceptron

clf3=Perceptron()
clf3.fit(xor_data.iloc[:,0:2],xor_data.iloc[:,-1])

print(clf3.coef_)
print(clf3.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values, clf=clf3, legend=2)
```
