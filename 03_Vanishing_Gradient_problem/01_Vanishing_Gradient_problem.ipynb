```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
import keras
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from keras.layers import Dense,Input
from keras.models import Sequential
```

```python
X,y = make_moons(n_samples=250, noise=0.05, random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1], c=y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model = Sequential()

model.add(Input(shape=(2,)))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs = 100)
```

```python
# so loss is not decreasing - hence Vanishhing Gradient problem is occuring
```

## Handling Gradient Descent by reducing the complexity of the model

```python
model2 = Sequential()

model2.add(Input(shape=(2,)))
model2.add(Dense(10,activation='sigmoid'))
model2.add(Dense(10,activation='sigmoid'))
model2.add(Dense(1, activation='sigmoid'))
```

```python
model2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model2.fit(X_train,y_train,epochs = 100)
```

## Handling Vanishing Gradient problem using relu activation function

```python
model3 = Sequential()

model3.add(Input(shape=(2,)))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(10,activation='relu'))
model3.add(Dense(1, activation='sigmoid'))
```

```python
model3.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model3.fit(X_train, y_train, epochs = 100)
```
