```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Input, Dense, Flatten
import numpy as np
import pandas as pd
```

```python
X = np.linspace(-10, 10, 100)
noise = np.random.randn(100) * 2  # Adding randomness
y = (3 * X + 10) + noise
```

```python
# scaling the data
X_norm = (X - X.mean()) / X.std()
```

```python
X_norm.shape
```

```python
model = Sequential()

model.add(Input(shape = (1,)))  #input layer
model.add(Flatten())

model.add(Dense(8,activation = 'relu')) #hidden layer 1
model.add(Dense(1,activation = 'linear')) #output layer
```

```python
model.compile(loss = 'mean_squared_error',optimizer = 'Adam',metrics = ['r2_score'])
```

```python
history = model.fit(X_norm,y, validation_split = 0.2, epochs = 1000) #as data is less so we need to perform more epochs
```
