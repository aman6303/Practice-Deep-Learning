```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'Price': [100, 102, 105, 103, 108, 115, 112, 118, 120, 125]
}
df = pd.DataFrame(data)
```

```python
df['EWMA_3(aplha=0.9)'] = df['Price'].ewm(alpha=0.9, adjust=False).mean()  #alpha tells how much weightage you want to give to each current point
df['EWMA_3(aplha=0.6)'] = df['Price'].ewm(alpha=0.6, adjust=False).mean()  # lesss alpha less weightage to current point -- more weightage to older values
df['EWMA_3(aplha=0.3)'] = df['Price'].ewm(alpha=0.3, adjust=False).mean()
```

```python
df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Price'], label='Actual Price', marker='o')
plt.plot(df['Date'], df['EWMA_3(aplha=0.9)'], label='EWMA (alpha=0.9)', color='red')
plt.plot(df['Date'], df['EWMA_3(aplha=0.6)'], label='EWMA (alpha=0.6)', color='green')
plt.plot(df['Date'], df['EWMA_3(aplha=0.3)'], label='EWMA (alpha=0.3)', color='black')
plt.legend()
plt.title('Price vs EWMA')
plt.show()
```
