```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Conv2D,Flatten, Input
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

# Valid padding

```python
# valid padding simply means no padding.
model = Sequential()

model.add(Input(shape=(28,28,1)))

model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

here we can clearly see the input size of image is decreasing

# Same padding

```python
model = Sequential()

# padding is applied in such a way that image size will remain same througout the each convolution
# keras calulate it automatically
model.add(Input(shape=(28,28,1)))

model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```
